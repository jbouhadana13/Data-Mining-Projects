Problem 3 (20 Points) Bagging for decision trees

Use ACT_07_Data for this part of programming assignment


DO K = 1 to 20;

1. STEP 01: Randomly select a sample of size 1,460 with replacement.
2. STEP 02: Build a decision trees model to predict the “HousePrice”
3. STEP 03: Use the model built in STEP 02 to score all 1,460 observation and keep all the predicted values f ̂_K (x) for each observation


END DO;

4. STEP 04: Calculate the bagging estimator using the formula f ̂_bag (x)=1/20 ∑_(K=1)^20▒〖f ̂_K (x) 〗

act 7 pt 5.JPG

5. STEP 05: Calculate the error estimator for the bagging estimator in “STEP 04” using the formula mse= [(∑_(i=1)^20▒(f ̂_K (x)-f ̂_bag (x))^2 )⁄20].

act 7 pt 6.JPG

6. Produce a box plot for all 20 bagging predictions with “predicted” “HousePrice” as the Y-axis.


```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
import numpy as np
import matplotlib.pyplot as plt

# Reading the CSV
data = pd.read_csv('ACT_07_Data.csv')

# Seperating the target variable (since we are creating a 4 loop, this has to be done first.)
X = data.drop('SalePrice', axis=1)
y = data['SalePrice']

# This is used to store the predicted values
predicted_values = np.empty((1460, 20))

# STEP THREE: Use the model built in STEP 02 to score all 1,460 observation and keep all the predicted values f ̂_K (x) for each observation


# The four loop is being created so that way when we do the 20 samples, it will automatically predict them


for i in range(20):
    # STEP ONE: Randomly select a sample of size 1,460 with replacement.
    sample = data.sample(n=1460, replace=True)
    new_X = sample.drop('SalePrice', axis=1)
    new_Y = sample['SalePrice']

    # STEP TWO: Build a decision trees model to predict the “HousePrice” (side note, this is Salesprice, and not Houseprice)
    dtr = DecisionTreeRegressor()
    dtr.fit(new_X, new_Y)

    # Use the model to predict the salesprice
    y_pred = dtr.predict(X)

    # This will help us store the predicted values
    predicted_values[:, i] = y_pred

# STEP FOUR: Calculate the bagging estimator
bagging = (1/20) * np.sum(predicted_values, axis=1)

# STEP FIVE: Calculate the error estimator for the bagging estimator in “STEP 04” using the formula mse= [(∑_(i=1)^20▒(f ̂_K (x)-f ̂_bag (x))^2 )⁄20].〗
mse = (1/20) * np.sum(np.square(predicted_values - bagging.reshape(-1, 1)))

#Adding the Bagging estimator and the predictive values to our dataset (we use concat because our data was printing from side to side not up and down)
df_pred = pd.DataFrame(predicted_values, columns=[f'K= {i+1}' for i in range(20)])
df_bag = pd.DataFrame(bagging, columns=['Bagging Estimator'])
print(pd.concat([df_pred, df_bag], axis=1))

# Printing the MSE
print(f'MSE: {mse}')
```

```python
#Step SIX Produce a box plot for all 20 bagging predictions with “predicted” “HousePrice” (SalePrice) as the Y-axis.

plt.boxplot(predicted_values)
plt.xticks(np.arange(1, 21), [f'K= {i+1}' for i in range(20)])
plt.ylabel('Predicted SalePrice')
plt.show()
```
